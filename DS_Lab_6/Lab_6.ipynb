```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('iris.csv')
df = data.copy()
data
```

```python
data.dtypes
```

```python
sns.set(style="darkgrid")

fig, axs = plt.subplots(2, 2, figsize=(7, 7))

sns.histplot(data=df, x="SepalLenCM",bins=10, color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="SepalWidthCM",bins=10, color="olive", ax=axs[0, 1])
sns.histplot(data=df, x="PetalLenCM",bins=10, color="gold", ax=axs[1, 0])
sns.histplot(data=df, x="PetalWidthCM",bins=10, color="teal", ax=axs[1, 1])

plt.show()
```

```python
sns.set(style="darkgrid")

fig, axs = plt.subplots(2, 2, figsize=(14, 14))

sns.boxplot(y="SepalLenCM",x="Species",data = df,ax=axs[0, 0])
sns.boxplot(y="SepalWidthCM",x="Species",data = df,ax=axs[0, 1])
sns.boxplot(y="PetalLenCM",x="Species",data = df,ax=axs[1, 0])
sns.boxplot(y="PetalWidthCM",x="Species",data = df,ax=axs[1, 1])

plt.show()
```

```python
sns.set(style="darkgrid")

fig, axs = plt.subplots(2, 2, figsize=(7, 7))

sns.distplot(df["SepalLenCM"],bins=10, color="skyblue", ax=axs[0, 0])
sns.distplot(df["SepalWidthCM"],bins=10, color="olive", ax=axs[0, 1])
sns.distplot(df["PetalLenCM"],bins=10, color="gold", ax=axs[1, 0])
sns.distplot(df["PetalWidthCM"],bins=10, color="teal", ax=axs[1,1])

plt.show()
```
